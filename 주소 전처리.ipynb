```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
df = pd.read_csv('서울시 대학 및 전문대학 DB 정보 (한국어).csv', encoding='cp949')
df.head()
```

```python
ad= df['주소']
print(ad.head())
```

```python
address = ad.copy()

for i in range(len(address)):
    a = address[i].split('(')
    address[i] = "".join(a[0])
print(address)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
#####Dataframe만들기######
address_df = pd.DataFrame({'키값': df['키값'],'연도':df['연도'],
                           '학교종류': df['학교종류'],'학교명':df['학교명'],
                           '본분교': df['본분교'],'학교상태':df['학교상태'],
                           '설립': df['설립'],'학교명':df['학교명'],
                           '주소':address,'위도':latitude,'경도':longitude,
                          '행정시': df['행정시'],'행정구':df['행정구'],'행정동':df['행정동']})
address_df

import pandas as pd
address_df.to_csv('df_address.csv')
```
